```python
# Specify the data columns we want to include (i.e. exclude Year, Month, Weekday Name)
data_columns = ['MAX_TEMPERATURE_C','MIN_TEMPERATURE_C']
# Resample to weekly frequency, aggregating with mean
tab_temp_weekly_mean = tab_temp[data_columns].resample('M').mean()
tab_temp_weekly_mean.head(3)
```

```python
# Start and end of the date range to extract
start, end = '2018-01', '2019-01'
# Plot daily and weekly resampled time series together
fig, ax = plt.subplots()
ax.plot(tab_temp.loc[start:end, 'MAX_TEMPERATURE_C'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(tab_temp_weekly_mean.loc[start:end, 'MAX_TEMPERATURE_C'],
marker='o', markersize=8, linestyle='-', label='Monthly Mean Resample')
ax.set_ylabel('MAX_TEMPERATURE_C (°C)')
ax.legend();
```

```python
# Compute the centered 7-day rolling mean
tab_temp_7d = tab_temp[data_columns].rolling(7, center=True).mean()
tab_temp_7d.head(10)
```

```python
# Start and end of the date range to extract
start, end = '2018-01', '2019-01'
# Plot daily, weekly resampled, and 7-day rolling mean time series together
fig, ax = plt.subplots()
ax.plot(tab_temp.loc[start:end, 'MAX_TEMPERATURE_C'],
marker='.', linestyle='-', linewidth=0.5, label='Daily')
ax.plot(tab_temp_weekly_mean.loc[start:end, 'MAX_TEMPERATURE_C'],
marker='o', markersize=8, linestyle='-', label='Weekly Mean Resample')
ax.plot(tab_temp_7d.loc[start:end, 'MAX_TEMPERATURE_C'],
marker='.', linestyle='-', label='7-d Rolling Mean')
ax.set_ylabel('Solar Production (GWh)')
ax.legend();
```

```python
# The min_periods=360 argument accounts for a few isolated missing days in the
# wind and solar production time series
tab_temp_365d = tab_temp[data_columns].rolling(window=365, center=True, min_periods=360).mean()
```

```python
# Plot daily, 7-day rolling mean, and 365-day rolling mean time series
fig, ax = plt.subplots()
ax.plot(tab_temp['MAX_TEMPERATURE_C'], marker='.', markersize=2, color='0.6',
linestyle='None', label='Daily')
ax.plot(tab_temp_7d['MAX_TEMPERATURE_C'], linewidth=2, label='7-d Rolling Mean')
ax.plot(tab_temp_365d['MAX_TEMPERATURE_C'], color='0.2', linewidth=3,
label='Trend (365-d Rolling Mean)')
# Set x-ticks to yearly interval and add legend and labels
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.legend()
ax.set_xlabel('Year')
ax.set_ylabel('Consumption (GWh)')
ax.set_title('Trends in Electricity Consumption');
```

```python
import re
future_pattern = re.compile("""([^,"]+|"[^"]+")(?=,|$)""")

def parseCSV(line):
     return future_pattern.findall(line)
```

```python
polluants = polluants_csv.map(parseCSV).filter(lambda x: x[0]!="id")
```

```python
polluants.take(1)
```

```python
def RepresentsInt(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False



def annee_mois(x):
    x2=list(x)
    if RepresentsInt(x[10]):
        return (x[13][0:4]+'_'+x[13][5:7],(x[8],x[10],x[1]))
    else:
        return (x[13][0:4]+'_'+x[13][5:7],(x[8],'-1',x[1]))
    
polluants_clean = polluants.map(lambda x: annee_mois(x))
polluants_clean.collect()
```

```python
temperature.take(1)
```

```python
## Calcul de la température moyenne mensuelle

def annee_mois_t(x):
    return (x[0][0:4]+'_'+x[0][5:7],(1,float(x[2]),float(x[1]),float(x[4]), \
                                    float(x[5]),float(x[6]), \
                                    float(x[3]),float(x[7]), \
                                    float(x[8]),float(x[9]), \
                                    float(x[10]),float(x[11])))

temperature_clean = temperature.map(lambda x: annee_mois_t(x)) \
            .filter(lambda x: ('2018' in x[0]) or ('2019' in x[0])) \
            .reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1],x[2]+y[2],x[3]+y[3],x[4]+y[4],x[5]+y[5],x[6]+y[6],x[7]+y[7],x[8]+y[8],x[9]+y[9],x[10]+y[10],x[11]+y[11])) \
            .mapValues(lambda x: (x[1]/x[0],x[2]/x[0],x[3]/x[0],x[4]/x[0],x[5]/x[0],x[6]/x[0],x[7]/x[0],x[8]/x[0],x[9]/x[0],x[10]/x[0],x[11]/x[0]))

#annee_mois , (min_temp(°C) 2,max_temp(°C) 1,température_matin(°C) 4,température_midi(°C) 5,température_soir(°C) 6,
#             windspeed(km/h) 3, précipitations(mm) 7, humidité (%) 8, visibilité (km) 9, pression (MB) 10, nuage(%) 11)
temperature_clean.collect()
```

```python
def tfloat(x):
    if x == '':
        return -1
    else:
        return float(x)

def clean_poll(x):
    if len(x) == 23:
        if x[10] == '5""':
            return (x[15].split(" ")[0],(x[0],1,tfloat(x[12]),x[9]+x[10],x[1]))
        else:
            return (x[15].split(" ")[0],(x[0],1,tfloat(x[12]),x[10],x[1]))
    elif len(x) == 21:
        return (x[13].split(" ")[0],(x[0],1,tfloat(x[10]),x[8],x[1]))
    else:
        return (x[14].split(" ")[0],(x[0],1,tfloat(x[11]),x[9],x[1]))



polluants_clean = polluants_j.map(lambda x: clean_poll(x)) \
            .filter(lambda x: x[1][2] != -1) \
            #.reduceByKey(lambda x,y: (x[0]+y[0],x[1]+y[1],x[2]+y[2],x[3]+y[3],x[4]+y[4],x[5]+y[5],x[6]+y[6],x[7]+y[7],x[8]+y[8],x[9]+y[9],x[10]+y[10],x[11]+y[11])) \
            #.mapValues(lambda x: (x[1]/x[0],x[2]/x[0],x[3]/x[0],x[4]/x[0],x[5]/x[0],x[6]/x[0],x[7]/x[0],x[8]/x[0],x[9]/x[0],x[10]/x[0],x[11]/x[0]))

```

```python
tab_poll = pd.read_csv('mesure_reunion_journaliere_cinor copie.csv',encoding='cp1252',header = 0)
tab_poll1 = tab_poll[tab_poll['nom_station'].isna()]
tab_poll2 = tab_poll[tab_poll['nom_station'].notna()]
tab_poll2.drop("influence",axis=1)
tab_poll2.columns
```

```python
tab_poll1 = tab_poll1['id'].str.split(',', expand=True)
tab_poll1.columns = tab_poll2.columns
```

```python
tab_pollution = pd.concat([tab_poll1, tab_poll2], axis=0, sort=False)
tab_pollution = tab_pollution[tab_pollution['statut_valid'] == 'A']
tab_pollution['DATE'] =tab_pollution['date_debut'].apply(lambda x: x.split(" ")[0])
tab_pollution = tab_pollution.astype({'valeur': numpy.float64})
```

```python
tab_pollution['DATE'] = pd.to_datetime(tab_pollution['DATE'])
tab_pollution = tab_pollution.set_index('DATE')
# Add columns with year, month, and weekday name
tab_pollution['Year'] = tab_pollution.index.year
tab_pollution['Month'] = tab_pollution.index.month
tab_pollution['Weekday Name'] = tab_pollution.index.weekday_name
tab_pollution
```

```python
Polluant_Regarde = "Dioxyde d'azote"
lieu_Regarde = "Lislet Geoffroy"
tab_PR = tab_pollution[(tab_pollution['nom_poll'] == Polluant_Regarde) & (tab_pollution['nom_station'] == lieu_Regarde)]
sns.set(rc={'figure.figsize':(11, 4)})
cols_plot = ['valeur']
tab_PR[cols_plot].plot(linewidth=0.5,figsize=(12, 25), subplots=True);
```

```python
# Specify the data columns we want to include (i.e. exclude Year, Month, Weekday Name)
data_columns = ['valeur']
# Resample to weekly frequency, aggregating with mean
tab_PR_weekly_mean = tab_PR[data_columns].resample('W').mean()
tab_PR_weekly_mean.head(3)
```
