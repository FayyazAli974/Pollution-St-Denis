## _Introduction_
Atmo Réunion est l’association, agréée par l’État au titre du code de l’Environnement (Art.L.220.3), pour la surveillance de la qualité de l’air à La Réunion. Elle met à disposition ses enregistrements de polluants captés grâce à des dispositifs situés sur la carte précédente.  

Le projet choisi a pour but la création d'un indicateur de pollution pour une journée donnée dans la ville de __Saint-Denis__, y connaissant les conditions climatiques de cette dernière.
Dans ce projet, nous n'allons pas faire de la prédiction de conditions climatiques, qui serait réalisé par Meteo-France par exemple. L'indicateur se baserait sur ces prévisions météos pour prévoir le seuil de polluant, et définir des actions à mener.

La ville de Saint-Denis est le chef lieu de La Réunion, et est soumise une forte circulation automobile (le parc automobile réunionnais est essentiellement Diesel), ainsi qu'à des embruns marins du fait de sa situation côtière. Après les scandales du diesel gate, ainsi que les interrogations actuels sur les cycles d'homologation, une surveillance du NO2 (dioxyte d'azote), des NOx et des particules est légitime.  
Pour l'instant, les mesures de circulation alternée n'existent pas à la Réunion, mais il faudra peut être y songer un jour, et ce type d'indicateur permettrait de décider quand l'appliquer.

De plus, l'Afrique du Sud par exemple est un pays pollueur à proximité de La Réunion, une étude de l'influence des vents et conditions climatiques pourrait ammener à regarder si les polluants de ce pays ne voyagent pas jusqu'à l'île.

La loi LAURE (loi sur l'air) du 30 Décembre 1996 fixe les seuils d'information de polluants:   
- dioxyde d'azote : 200 µg/m3  
- NOx : 400 µg/m3  
- dioxyde de soufre : 300 µg/m3  
- particules fines : 80 µg/m3  
- ozone : 240 µg/m3

Les données de conditions climatiques passés sur la ville de saint-denis proviennent du site historique-meteo.net. Nous avons accès aux paramètres journaliers suivants:
- Température maximale (°C)
- Température minimale (°C)
- Vitesse maximale du vent (km/h)
- Température matin (°C)
- Température midi (°C)
- Température soir (°C)
- Précipitations (mm)
- Humidité maximale (%)
- Visibilité moyenne (km)
- Pression Max (MB)
- Couverture nuageuse moyenne (%)

Il faudra faire attention aux corrélations de variables dans ces données, notamment les différentes températures.


Aprés analyse des données, il s'avère que les niveaux de polluants mesurés sur la ville de Saint-Denis sont (heureusement) très loin des seuils d'alerte, nous allons donc pour cette étude faire deux modèles:
- __Classification:__ on fixe un seuil arbitraire d'alerte puis on prédit si pour une nouvelle journée (dont on connaît les prédictions météorologiques), celle-ci sera à risque ou non.  

- __Régression:__ on prédit une valeur de concentration de polluant à partir des conditions météorologiques.

### <font color=Green> 2.1 Classification <font color>

```python
pt_poll = pt.filter(lambda x : (x[1][0][3] == Polluant_Regarde) & (x[1][0][4] == lieu_Regarde))
```

```python
def detect_seuil(x):
    if x[1][0][2]<Seuil_Polluant:
        return 0
    else:
        return 1
Poll_Seuil = pt.map(lambda x:detect_seuil(x))
```

Nous allons prendre 80% de l'échantillon pour le train, et 20% pour le test

```python
weights = [0.8, 0.2]
seed = 42

pt_train, pt_test = pt_poll.randomSplit(weights, seed)
```

```python
# (date, (1, T Min, T Max, T Morning, T Noon, T Evening, Windspeed, Rain, Humidity_Max, Visibility, Pressure, Cloud))
Temp_Min = pt_train.map(lambda x: x[1][1][1]).distinct().collect()
Temp_Max = pt_train.map(lambda x: x[1][1][2]).distinct().collect()
Temp_Morning = pt_train.map(lambda x: x[1][1][3]).distinct().collect()
Temp_Noon = pt_train.map(lambda x: x[1][1][4]).distinct().collect()
Temp_Evening = pt_train.map(lambda x: x[1][1][5]).distinct().collect()
Windspeed = pt_train.map(lambda x: x[1][1][6]).distinct().collect()
Rain = pt_train.map(lambda x: x[1][1][7]).distinct().collect()
Humidity_Max = pt_train.map(lambda x: x[1][1][8]).distinct().collect()
Visibility = pt_train.map(lambda x: x[1][1][9]).distinct().collect()
Pressure = pt_train.map(lambda x: x[1][1][10]).distinct().collect()
Cloud = pt_train.map(lambda x: x[1][1][11]).distinct().collect()
```

```python
from pyspark.mllib.regression import LabeledPoint
from numpy import array

def create_labeled_point(row):
    # convert label to binary label
    
    p_seuil = 0
    if detect_seuil(row):
        p_seuil = 1
        
    return LabeledPoint(p_seuil, array([float(x) for x in row[1][1][1:]]))

training_data = pt_train.map(create_labeled_point)
test_data = pt_test.map(create_labeled_point)
all_data = pt_poll.map(create_labeled_point)
```

```python
from pyspark.mllib.tree import DecisionTree, DecisionTreeModel
from time import time

# Build the model
tree_model = DecisionTree.trainClassifier(training_data, numClasses=2, categoricalFeaturesInfo={},impurity='gini',
                                          maxDepth=20, maxBins=300)
```

```python
predictions = tree_model.predict(test_data.map(lambda p: p.features))
labels_and_preds = test_data.map(lambda p: p.label).zip(predictions)
```

### Recherche d'un meilleur classifieur: passage en DataFrame pour bénéficier de la libraire ML

```python
from pyspark.sql.types import StructType, StructField 
from pyspark.sql.types import DoubleType, IntegerType 
from pyspark.ml.feature import VectorAssembler
from pyspark.sql import SQLContext
from pyspark.ml.linalg import VectorUDT,Vectors
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression
from pyspark.ml.tuning import ParamGridBuilder, TrainValidationSplit
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
```

```python
def toFloat(x):
    if x == '?':
        return 5.0
    else:
        return float(x)


def do_line(row):
    p_seuil = 0
    if detect_seuil(row):
        p_seuil = 1
    return (Vectors.dense([toFloat(e) for e in row[1][1][1:]]),p_seuil)
```

```python
# Transformation en RDD pour utiliser la librairie ML
schema = StructType([StructField("features", VectorUDT(), True),StructField("label",IntegerType(),True)])

pt_train_rdd = pt_train.map(do_line)
pt_train_df = pt_train_rdd.toDF(schema)

pt_test_rdd = pt_test.map(do_line)
pt_test_df = pt_test_rdd.toDF(schema)

all_data_rdd = pt_poll.map(do_line)
all_data_df = all_data_rdd.toDF(schema)
```

```python
pt_train_df.take(5)
```

- Train - Validation Split:

```python
tree_model = DecisionTreeClassifier(labelCol="label", featuresCol="features")

paramGrid = ParamGridBuilder().addGrid(tree_model.maxDepth, [i for i in range(2,20,4)])\
                              .addGrid(tree_model.maxBins, [i for i in range(10,300,20)])\
                              .build()

tvs = TrainValidationSplit(estimator=tree_model,\
                           estimatorParamMaps=paramGrid,\
                           evaluator=BinaryClassificationEvaluator(),\
                           trainRatio=0.8)\

model_besttrain2 = tvs.fit(pt_train_df)
```

```python
prediction_besttrain2 = model_besttrain2.transform(pt_test_df)
result_besttrain2 = BinaryClassificationEvaluator(labelCol="label",metricName='areaUnderROC').evaluate(prediction_besttrain2)
# Train validation give us this ROC curve:
result_besttrain2
```

```python
total = prediction_besttrain2.count()
tp_total = prediction_besttrain2.where(prediction_besttrain2.label == 1).count()
tp = prediction_besttrain2.where(prediction_besttrain2.label == 1).where(prediction_besttrain2.prediction == 1).count()
tn = prediction_besttrain2.where(prediction_besttrain2.label == 0).where(prediction_besttrain2.prediction == 0).count()
accuracy = (tp+tn)/total
accuracy
```

On retrouve exactement la même accuracy qu'en utilisant la librairie RDD, c'est certainement exactement le même arbre que précédemment

- Cross Validation:

```python
tree_model = DecisionTreeClassifier(labelCol="label", featuresCol="features")

crossval = CrossValidator(estimator=tree_model,
                          estimatorParamMaps=paramGrid,
                          evaluator=BinaryClassificationEvaluator(),
                          numFolds=3)  

cvModel_tree = crossval.fit(pt_train_df)
```

```python
prediction_crosstrain_tree = cvModel_tree.transform(pt_test_df)
result_cv = BinaryClassificationEvaluator(labelCol="label",metricName='areaUnderROC').evaluate(prediction_crosstrain_tree)
# Cross validation give us this ROC curve:
result_cv
```

```python
total = prediction_crosstrain_tree.count()
tp_total = prediction_crosstrain_tree.where(prediction_crosstrain_tree.label == 1).count()
tp = prediction_crosstrain_tree.where(prediction_crosstrain_tree.label == 1).where(prediction_crosstrain_tree.prediction == 1).count()
tn = prediction_crosstrain_tree.where(prediction_crosstrain_tree.label == 0).where(prediction_crosstrain_tree.prediction == 0).count()
accuracy = (tp+tn)/total
accuracy
```

On a trouvé un meilleur modèle par cross-validation
